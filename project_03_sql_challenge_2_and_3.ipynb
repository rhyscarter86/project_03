```python
query = '''
SELECT match.id, away_team_goal, home_team_goal, (away_team_goal+home_team_goal) as total_goals
FROM Match
WHERE away_team_goal = home_team_goal
ORDER BY total_goals DESC
;
'''

# Creating dataframe from query.
soccer_ties_median = conn.execute(query).fetchall()
soccer_ties_median_df = pd.DataFrame(soccer_ties_median)
```

```python
query = '''
CREATE TABLE tie_goals AS
SELECT match.id, away_team_goal, home_team_goal, (away_team_goal+home_team_goal) as total_goals
FROM Match
WHERE away_team_goal = home_team_goal
ORDER BY total_goals DESC
;
'''

# Creating dataframe from query.
soccer_ties_median = conn.execute(query).fetchall()
soccer_ties_median_df = pd.DataFrame(soccer_ties_median)
```

```python
query = '''

SELECT AVG(total_goals)
FROM (SELECT total_goals
      FROM match
      ORDER BY total_goals
      LIMIT 2 - (SELECT COUNT(*) FROM match) % 2    -- odd 1, even 2
      OFFSET (SELECT (COUNT(*) - 1) / 2
              FROM match))
;
'''

# Creating dataframe from query.
soccer_ties_median = conn.execute(query).fetchall()
soccer_ties_median_df = pd.DataFrame(soccer_ties_median)
```
